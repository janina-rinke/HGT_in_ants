```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd

# Data was obtained using: "LGTs.CDSLevel.UniProt.Info.Reannotated" table.
# Make data: I have 5 groups and 11 subgroups
group_names=['Alphaproteobacteria', 'Betaproteobacteria', 'Gammaproteobacteria','Mollicutes', 'Sphingobacteriia']
group_size=[880,1,91,68,11]
subgroup_names=['Wolbachia', 'Other', 'Paraburkholderia', 'Blochmannia', 'Sodalis', 'Yersinia', 'Other', 
'Spiroplasma', 'Mycoplasma', 'Cardinium']
subgroup_size=[874,6,1,11,28,6,46,64,4,11]

# Create colors
a, b, c, d, e=[plt.cm.YlGnBu, plt.cm.summer, plt.cm.YlGn, plt.cm.twilight, plt.cm.PuBuGn]

font_color = 'black'
hfont = {'fontname':'Calibri'}

# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors= 
[a(0.9), b(0.6), c(0.8), d(0.1), e(0.4)])
plt.setp( mypie, width=0.3, edgecolor='white')

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, 
labels=subgroup_names, labeldistance=None, colors=[a(0.6), a(0.2), 
b(0.4), b(0.3), c(0.4), c(0.3), c(0.7), c(0.9), d(0.4), d(0.3), e(0.3)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)

subgroup_names_legs=['Wolbachia (Alphaproteobacteria)', 'Other (Alphaproteobacteria)', 'Paraburkholderia (Betaproteobacteria)', 'Blochmannia (Gammaproteobacteria)', 'Sodalis (Gammaproteobacteria)', 'Yersinia (Gammaproteobacteria)', 'Other (Gammaproteobacteria)', 
'Spiroplasma (Mollicutes)', 'Mycoplasma (Mollicutes)', 'Cardinium (Sphingobacteria)']
plt.legend(loc=(0.9, 0.1))
handles, labels = ax.get_legend_handles_labels()

ax.legend(handles[5:],subgroup_names_legs, loc=(1.3, 0.1))

plt.savefig('Bacterial_distribution_taxonomy.pdf', bbox_inches='tight', dpi=192)
plt.show()
```
